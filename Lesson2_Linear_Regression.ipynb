```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_data = [1,2,3]
y_data = [2,4,6]
```

```python
def forward(x,w):
    return x * w
```

```python
def loss(x,y):
    y_pred = forward(x,w)
    return (y_pred - y)**2
```

```python
w_list = []
mse_list = []
```

Let's guess for each w and find it's loss

```python
for w in np.arange(0.1,4,0.1):
    l_sum = 0
    for x,y in zip(x_data,y_data):
        y_pred_val = forward(x,w)
        l = loss(x,y)
        l_sum += l
    w_list.append(w)
    mse_list.append(l_sum/3)
    
```

```python
plt.plot(w_list,mse_list)
plt.xlabel("w")
plt.ylabel("loss")
```
